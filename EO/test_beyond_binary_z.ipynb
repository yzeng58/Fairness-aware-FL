# Simulations

```python
acc, EO = np.zeros(10), np.zeros(10)
acc_p, EO_p = np.zeros(10), np.zeros(10)
for seed in range(10):
    print("| seed: %d |" % seed)
    server = Server(logReg(num_features=3, num_classes=2, seed = seed), synthetic_info, train_prn = False, seed = seed, ret = True, Z = 3)
    acc[seed], EO[seed] = server.Unconstrained(num_rounds = 10, local_epochs = 30, optimizer = 'adam', learning_rate = 0.0005)
    acc_p[seed], EO_p[seed] = server.ThresholdAdjust(num_rounds = 10, local_epochs = 30, epsilon = 0.02, learning_rate = 0.005)
    print("|------ Acc.: %.3f | EO Dis.: %.3f |" % (acc[seed], EO[seed]))
    print("|Post-processing ------ Acc.: %.3f | EO Dis.: %.3f |" % (acc_p[seed], EO_p[seed]))
print("| Acc.: %.3f(%.3f) | EO Dis.: %.3f(%.3f) |" % (acc.mean(), acc.std(), EO.mean(), EO.std()))
print("| Post-processing Acc.: %.3f(%.3f) | EO Dis.: %.3f(%.3f) |" % (acc_p.mean(), acc_p.std(), EO_p.mean(), EO_p.std()))
```
