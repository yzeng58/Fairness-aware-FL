```python
acc, eo = np.zeros(10), np.zeros(10)
for seed in range(10):
    print("| seed: %d |" % seed)
    server = Server(logReg(num_features=compas_num_features, num_classes=2, seed = seed), compas_info, train_prn = False, seed = seed, ret = True)
    acc[seed], eo[seed] = server.Unconstrained(num_rounds = 10, local_epochs = 20, optimizer = 'adam', learning_rate = 0.002)
    print("|------ Acc.: %.3f | EO Dis.: %.3f |" % (acc[seed], eo[seed]))
print("| Acc.: %.3f(%.3f) | EO Dis.: %.3f(%.3f) |" % (acc.mean(), acc.std(), eo.mean(), eo.std()))
```
